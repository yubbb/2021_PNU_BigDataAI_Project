```python
import pandas as pd
import numpy as np
import datetime
import time
from datetime import timedelta
```

```python
df = pd.read_csv('data/대선황교안_정제.csv')
df
```

```python
df.info()
```

```python
# datetime으로 형변환
df['Comment_date'] = df['Comment_date'].apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d'))
df
```

```python
# 2021년 3월 1일 이후의 데이터만 추출
df = df.query('Comment_date >= datetime.datetime(2021,3,1)')
df
```

```python
df.to_csv('data/대선황교안.csv',encoding='utf-8-sig', index=False)
```
