## 오세훈 (야후 파이낸스 사용)

```python
sh = ['진양산업','진양홀딩스','진양화학'] # 진양폴리 X

# 오세훈후보 관련주 시각화
for search in sh :
    # 회사명으로 주식 종목 코드를 획득할 수 있도록 하는 함수
    def get_code(df, name):
        code = df.query("name=='{}'".format(name))['code'].to_string(index=False)

        # 위와같이 code명을 가져오면 앞에 공백이 붙어있는 상황이 발생하여 앞뒤로 sript() 하여 공백 제거
        code = code.strip()
        return code

    # excel 파일을 다운로드하는거와 동시에 pandas에 load하기
    # 흔히 사용하는 df라는 변수는 data frame을 의미합니다.
    code_df = pd.read_html('http://kind.krx.co.kr/corpgeneral/corpList.do?method=download', header=0)[0]

    # data frame정리
    code_df = code_df[['회사명', '종목코드']]

    # data frame title 변경 '회사명' = name, 종목코드 = 'code'
    code_df = code_df.rename(columns={'회사명': 'name', '종목코드': 'code'})

    # 종목코드는 6자리로 구분되기때문에 0을 채워 6자리로 변경
    code_df.code = code_df.code.map('{:06d}'.format)

    # 코드 구하기
    code = get_code(code_df, search)

    # yahoo의 주식 데이터 종목은 코스피는 .KS, 코스닥은 .KQ가 붙는다
    code = code + '.KS'
    # code = code + '.KQ'

    # get_data_yahoo API를 통해서 yahoo finance의 주식 종목 데이터(단일화 이후)를 가져온다
    # 등락률 구하기 위해 22일 전 데이터를 가져와야 하는데 주말이라 19일 들고옴
    df = data.get_data_yahoo(code,start='2021-03-19')
    # print(code)
    
    # 단일화 이후 ~ 선거일 까지의 날짜만 들고온 후 종가 기준으로 시각화
    df = df.query('Date <= datetime.datetime(2021,4,10)')
    print(f'{search} 주가현황')
    display(df)
    
    # seaborn lineplot 그리기
    sns.lineplot(data=df['Close'])

    # title 설정
    plt.title(f'{search} 주식 종가',
                      fontsize = 'x-large',
                      fontweight = 'bold')
    
    # 4월 7일 확인
    plt.axvline(x=datetime.datetime(2021, 4, 7), color='r',linestyle=':', linewidth=3)

    # xticks 45도 회전
    plt.xticks(rotation=45)

    plt.show()
```

## 오세훈 (야후 파이낸스 사용)

```python
sh = ['진양산업','진양홀딩스','진양화학'] # 진양폴리 X
total = []

# 오세훈후보 관련주 시각화
for search in sh :
    # 회사명으로 주식 종목 코드를 획득할 수 있도록 하는 함수
    def get_code(df, name):
        code = df.query("name=='{}'".format(name))['code'].to_string(index=False)

        # 위와같이 code명을 가져오면 앞에 공백이 붙어있는 상황이 발생하여 앞뒤로 sript() 하여 공백 제거
        code = code.strip()
        return code

    # excel 파일을 다운로드하는거와 동시에 pandas에 load하기
    # 흔히 사용하는 df라는 변수는 data frame을 의미합니다.
    code_df = pd.read_html('http://kind.krx.co.kr/corpgeneral/corpList.do?method=download', header=0)[0]

    # data frame정리
    code_df = code_df[['회사명', '종목코드']]

    # data frame title 변경 '회사명' = name, 종목코드 = 'code'
    code_df = code_df.rename(columns={'회사명': 'name', '종목코드': 'code'})

    # 종목코드는 6자리로 구분되기때문에 0을 채워 6자리로 변경
    code_df.code = code_df.code.map('{:06d}'.format)

    # 코드 구하기
    code = get_code(code_df, search)

    # yahoo의 주식 데이터 종목은 코스피는 .KS, 코스닥은 .KQ가 붙는다
    code = code + '.KS'
    # code = code + '.KQ'

    # get_data_yahoo API를 통해서 yahoo finance의 주식 종목 데이터(단일화 이후)를 가져온다
    # 등락률 구하기 위해 22일 전 데이터를 가져와야 하는데 주말이라 19일 들고옴
    df = data.get_data_yahoo(code,start='2021-03-19')
    # print(code)
    
    # 단일화 이후 ~ 선거일 까지의 날짜만 들고온 후 종가 기준으로 시각화
    df = df.query('Date <= datetime.datetime(2021,4,10)')
    
    # 등락률 계산
    df_close = df['Close']
    rate = [0]
    for i in range(len(df_close)-1):
        rate.append(round(((df_close[i+1]-df_close[i])/df_close[i])*100,2))
    df['등락률']=rate
    
    # 등락률 계산위해 가져왔던 2021-03-19 제거
    df = df.iloc[1:,:]
    
    # 평균 구하기 위해 total에 append
    total.append(df['등락률'])
    
    print(f'{search} 주가현황')
    display(df)

    # seaborn lineplot 그리기
    sns.lineplot(data=df['등락률'])

# --------------------------------------------------------------------------------------------------------------   
# 진양폴리는 따로 구하기

# 시작날짜
start = datetime.datetime(2021, 3, 19) # 등락률 구하기 위해 22일 전 데이터를 가져와야 하는데 주말이라 19일 들고옴
# 종료날짜
end = datetime.datetime(2021, 4, 10)

# 진양폴리 주식 데이터를 가져온다.(직접 코드 찾은 후 입력)
df = data.DataReader('010640.KS', 'yahoo', start, end)
df_close = df['Close']
rate = [0]
for i in range(len(df_close)-1):
    rate.append(round(((df_close[i+1]-df_close[i])/df_close[i])*100,2))
df['등락률']=rate

# 등락률 계산위해 가져왔던 2021-03-19 제거
df = df.iloc[1:,:]
# 평균 구하기 위해 total에 append
total.append(df['등락률'])

print('진양폴리 주가현황')
display(df)

# seaborn lineplot 그리기
sns.lineplot(data=df['등락률'])

# ---------------------------------------------------------------------------------------------------------------
# 관련주 4개의 평균 구하기
avg = []
num = 0
for i in range(len(df)):
    for j in range(4):
        num+=total[j][i]
    avg.append(round(num/4,2))
    num = 0
avg

plt.plot(df.index,avg,color='black',marker='o',linestyle='dashed',linewidth=3)
# ---------------------------------------------------------------------------------------------------------------
# 시각화
# title 설정 
plt.title(f'오세훈 후보 관련주 등락률',
                  fontsize = 'x-large',
                  fontweight = 'bold')

# 4월 7일 확인
plt.axvline(x=datetime.datetime(2021,4,7), color='red',linestyle=':', linewidth=3)
#plt.text('재보궐선거일 --->',color='red', verticalalignment='center',horizontalalignment='center') # 오류 why?

# xticks 45도 회전
plt.xticks(rotation=45)

plt.ylim([-30,30])
plt.legend(sh+['평균'])
plt.show()
```
