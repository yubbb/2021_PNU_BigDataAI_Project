## 이재명 관련주

```python
jm = ['CS','에이텍','에이텍티앤','동신건설']
total = []

for i in range(4):
    # 이재명 관련주 주가 들고오기
    df = pd.read_csv(f'data/{jm[i]}.csv',encoding='cp949')
    #print(f'{jm[i]} 주가 추이')
    #display(df)
    
    # 평균 구하기 위해 total에 append
    total.append(df['등락률'])

    sns.lineplot(x=df['일자'],y=df['등락률'])

# 관련주 4개의 평균 구하기
avg = []
num = 0
for i in range(len(df)):
    for j in range(4):
        num+=total[j][i]
    avg.append(round(num/4,2))
    num = 0
avg

plt.plot(df['일자'],avg,color='black',linestyle='dashed',linewidth=3)

# title 설정
# plt.title(f'윤석열 후보 관련주 등락률',
#                   fontsize = 'x-large',
#                   fontweight = 'bold')

# 4월 7일 확인
plt.axhline(y=0, color='black', linewidth=1)

# xticks 45도 회전
plt.xticks(rotation=45)

plt.ylim([-20,20])
plt.legend(jm+['평균'])
plt.grid()
plt.show()
```
