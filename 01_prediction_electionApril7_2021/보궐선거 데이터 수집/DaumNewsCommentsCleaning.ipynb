```python
import pandas as pd
```

```python
df_chosun = pd.read_csv('chosun_comment.csv')
df_joongang = pd.read_csv('joongang_comment.csv')
df_donga = pd.read_csv('donga_comment.csv')
df_jtbc = pd.read_csv('jtbc_comment.csv')
df_khan = pd.read_csv('khan_comment.csv')
```

```python
df_total = pd.concat((df_chosun, df_joongang, df_donga, df_jtbc, df_khan))
df_total = df_total.reset_index(drop=True)
```

```python
## 제목 한자 처리
# 한자를 한글로 바꾸는 함수
def exchange_hanja(string):
    
    import re

    string = re.sub('文', '문재인', string)   
    string = re.sub('與', '더불어민주당', string) 
    string = re.sub('野', '야당', string) 
    string = re.sub('法', '법', string) 
    string = re.sub('眞', '진', string) 
    string = re.sub('尹', '윤석열', string) 
    string = re.sub('朴', '박영선', string) 
    string = re.sub('前', '전', string) 
    string = re.sub('靑', '청와대', string) 
    string = re.sub('檢', '검찰', string) 
    string = re.sub('親', '친', string) 
    string = re.sub('兆', '조', string) 
    string = re.sub('發', '발', string) 

    string = re.sub('安', '안철수', string) 
    string = re.sub('反', '반', string) 
    string = re.sub('巨', '거', string) 
    string = re.sub('聯', '연', string) 
    string = re.sub('政', '정', string) 
    string = re.sub('韓', '한국', string) 
    string = re.sub('美', '미국', string) 
    string = re.sub('後', '후', string) 
    string = re.sub('正', '정', string) 
    string = re.sub('道', '도', string) 
    string = re.sub('族', '족', string) 
    string = re.sub('琴', '금태섭', string) 
    string = re.sub('朝', '조', string) 

    string = re.sub('鮮', '선', string) 
    string = re.sub('心', '심', string) 
    string = re.sub('日', '일본', string) 
    string = re.sub('票', '표', string) 
    string = re.sub('北', '북한', string) 
    string = re.sub('軍', '군', string) 
    string = re.sub('部', '부', string) 
    string = re.sub('案', '안', string) 
    string = re.sub('脫', '탈', string) 
    string = re.sub('行', '행', string) 
    string = re.sub('吳', '오세훈', string) 
    string = re.sub('黨', '당', string) 
    string = re.sub('病', '병', string) 

    string = re.sub('丁', '정세균', string) 
    string = re.sub('性', '성', string) 
    string = re.sub('秋', '추미애', string) 
    string = re.sub('先', '선', string) 
    string = re.sub('盧', '노무현', string) 
    string = re.sub('禹', '우상호', string) 
    string = re.sub('女', '여', string) 
    string = re.sub('羅', '나경원', string) 
    string = re.sub('史', '사', string) 
    string = re.sub('中', '중국', string) 
    string = re.sub('勢', '세력', string) 
    string = re.sub('非', '비', string) 
    string = re.sub('無', '무', string) 

    string = re.sub('金', '김', string) 
    string = re.sub('亞', '아시아', string) 
    string = re.sub('說', '설', string) 
    string = re.sub('對', '대', string) 
    string = re.sub('話', '화', string) 
    string = re.sub('康', '강', string) 
    string = re.sub('洪', '홍남기', string) 
    string = re.sub('天', '천', string) 
    string = re.sub('運', '운', string) 
    string = re.sub('卞', '변창흠', string) 
    string = re.sub('路', '로', string) 
    string = re.sub('年', '연', string) 

    return string
```

```python
df_total['제목'] = [exchange_hanja(title) for title in df_total['제목']]
```

```python
## 라벨링 칼럼 만들기
df_total['지역(서울:1, 부산:2)'] = ''
df_total['정당(1:더불어민주당,2:국민의힘)'] = ''
df_total['정당평가(부정;0, 긍정:1)'] = ''
df_total['후보(기호 순)'] = ''
df_total['후보평가(부정;0, 긍정:1)'] = ''
df_total['제목댓글'] = [title + ' ' + comment for title, comment in zip(df_total['제목'], df_total['댓글'])]
df_total
```

```python
df_total.to_csv('재보궐선거댓글데이터_최종.csv', encoding='utf-8-sig', index=False)
```
