```python
import numpy as np
import pandas as pd
```

# 데이터 불러오기

```python
key = pd.read_csv('data/keyword.csv',index_col='Unnamed: 0')
key
```

```python
pledge = pd.read_csv('data/pledge.csv',index_col='Unnamed: 0')
pledge
```

# 후보공약의 keyword 개수만큼 keyword_count 증가

```python
# count list 생성
count = [0]*len(pledge)

# 후보공약에 키워드가 있으면 count 증가
for i in range(len(pledge['pledge'])):
    # print('---'*30)
    for j in range(len(key['keyword'])):
        # print(key['keyword'].iloc[j])
        if key['keyword'].iloc[j] in pledge['pledge'].iloc[i]:
            print(f"포함 키워드 : {key['keyword'].iloc[j]}")
            print(f"공약 : {pledge['pledge'].iloc[i]}")
            count[i] += 1
count
```

# 후보공약에 keyword_count 컬럼 추가한 df 생성

```python
df = pledge
df['keyword_count'] = count
df
```

# keyword_count 높은 순으로 보기

```python
df.groupby('name').sum().sort_values(by='keyword_count',ascending=False)
```
